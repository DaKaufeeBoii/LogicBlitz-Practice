```python
class ATM:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        print(f"Welcome {self.name}! Your account has been created with ₹{self.balance}.")

    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited. New balance: ₹{self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn. Remaining balance: ₹{self.balance}")
        else:
            print("Insufficient balance!")

    def check_balance(self):
        print(f"{self.name}, your current balance is ₹{self.balance}")

# Create objects (constructor is called automatically)
user1 = ATM("Sai Krishna", 1000)
user1.deposit(500)
user1.withdraw(300)
user1.check_balance()
```

```python
class ATM:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        print(f"Account created for {self.name} with balance ₹{self.balance}")

    def check_balance(self):
        print(f"{self.name}, your current balance is ₹{self.balance}")

    def __del__(self):
        print(f"Account of {self.name} is now closed. Goodbye!")

# Create object
user1 = ATM("Sai Krishna", 1000)

# Use method
user1.check_balance()

# Delete object manually (to trigger destructor)
del user1
```

```python
# class method
class School:
    school_name = "Edunet High School"  # class attribute
    def __init__(self,name):
        self.name = name

    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name
        print(f"School name changed to: {cls.school_name}")
    def show_details(self):
        print(f"{self.name},{self.school_name}")
```

```python
s1 = School("kl")
s1.show_details()

s1.change_school("xyz")
s1.show_details()
```

```python
## encapsulation

class BankAccount:
    def __init__(self, name, balance):
        self.__name = name        # private attribute
        self.__balance = balance  # private attribute

    # Getter method → to access private data
    def get_balance(self):
        return self.__balance

    # Setter method → to modify private data
    def set_balance(self, new_balance):
        if new_balance >= 0:
            self.__balance = new_balance
            print("Balance updated successfully!")
        else:
            print("Balance cannot be negative!")

# create object
user1 = BankAccount("Sai Krishna", 1000)

# using getter
print("Current Balance:", user1.get_balance())

# using setter
user1.set_balance(2000)

# verify change
print("Updated Balance:", user1.get_balance())

# trying to set negative value
user1.set_balance(-500)
```

```python
# Parent class
class Person:
    def show_details(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Child class (inherits from Person)
class Student(Person):
    def __init__(self, name, age, student_id):
        self.name = name
        self.age = age
        self.student_id = student_id

    def show_student(self):
        print(f"Student ID: {self.student_id}")

# Create an object of child class
s1 = Student("Sai Krishna", 25, "ST101")

# Access methods
s1.show_details()   # from parent
s1.show_student()   # from child
```

```python
# Parent class 1
class Engine:
    def engine_info(self):
        print("Engine: 1500cc Turbo Petrol Engine")

# Parent class 2
class MusicSystem:
    def music_info(self):
        print("Music System: Sony Surround Sound")

# Child class inheriting from both
class Car(Engine, MusicSystem):
    def car_info(self):
        print("Car: Model - Honda City VX, Year - 2025")

# Create an object of the child class
my_car = Car()

# Access methods from both parents and the child
my_car.car_info()
my_car.engine_info()
my_car.music_info()
```

```python
# Base class #multiple inheritance
class Vehicle:
    def info(self):
        print("This is a vehicle.")

# Intermediate class
class Car(Vehicle):
    def car_info(self):
        print("Car type: Sedan")

# Derived class
class ElectricCar(Car):
    def battery_info(self):
        print("Battery capacity: 80 kWh")

# Create object of child class
tesla = ElectricCar()

# Access methods from all levels
tesla.info()         # from Vehicle
tesla.car_info()     # from Car
tesla.battery_info() # from ElectricCar
```

```python
# Parent class
class Vehicle:
    def vehicle_type(self):
        print("This is a vehicle.")

# Child class 1
class Car(Vehicle):
    def car_info(self):
        print("Car type: Sedan")

# Child class 2
class Bike(Vehicle):
    def bike_info(self):
        print("Bike type: Sports")

# Create objects of child classes
my_car = Car()
my_bike = Bike()

# Access parent and child methods
my_car.vehicle_type()  # from Vehicle
my_car.car_info()      # from Car

my_bike.vehicle_type() # from Vehicle
my_bike.bike_info()    # from Bike
```

```python
## method overriding
# Parent class
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

# Child class 1
class Car(Vehicle):
    def start(self):
        print("Car engine roars to life!")
```

```python
from abc import ABC, abstractmethod

# Abstract class
class ATM(ABC):

    @abstractmethod
    def withdraw(self):
        pass

    @abstractmethod
    def deposit(self):
        pass

# Concrete class (child class)
class SBI_ATM(ATM):

    def withdraw(self):
        print("Cash withdrawn successfully!")

    def deposit(self):
        print("Cash deposited successfully!")

# Creating object
atm1 = SBI_ATM()
atm1.deposit()
atm1.withdraw()
```
